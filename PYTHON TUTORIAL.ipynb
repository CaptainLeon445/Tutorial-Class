### for loops and range function ()

for loops consist the following:
- the for keyword
- a variable name
- the in keyword
- a call to the range() function
- a colon
- indented code block of for clause

syntax:
for variable in range(start, stop, step):
    code block of for clause
    
- start argument is the starting number of a sequence i.e the lower limit. By default it starts with zero
- stop argument is the upper limit. The upper limit number is not always included in the result
- step is the differnce between each number in the result. By the default, it is 1 if not specified

ways you can call your range():

- range(stop)
- range(start, stop)
- range (start, stop, step)

### QUIZ APP

```python
print("QUIZ APP")

score=0
Name=input("Enter your Name: ")

while True:
    
    print('''
    1. What is the name of President of Nigeria?
    
    a. Mohammed Buhari
    b. Lamido Sanusi
    c. Elon Musk
    d. Bill Gates
    ''')
    
    ans=input("Option: ")
    ans=ans.lower()
    correct_ans='a'
    if ans==correct_ans:
        score+=1
    else:
        score
        
    print('''
    2. Which of the following colors is in Nigeria Flag?
    
    a. red
    b. white
    c. blue
    d. yellow
    ''')
    
    ans=input("Option: ")
    ans=ans.lower()
    correct_ans='b'
    if ans==correct_ans:
        score+=1
    else:
        score
        
    print('''
    3. Which of the following is not an Animal?
    
    a. Lion
    b. Goat
    c. Paper
    d. Fish
    ''')
    
    ans=input("Option: ")
    ans=ans.lower()
    correct_ans='c'
    if ans==correct_ans:
        score+=1
    else:
        score=0
        
    print('''
    4. Which of the following is not a color?
    
    a. red
    b. blue
    c. white
    d. blood
    ''')
    
    ans=input("Option: ")
    ans=ans.lower()
    correct_ans='d'
    if ans==correct_ans:
        score+=1
    else:
        score=0
        
    break
print("Name: ",Name)
print("Score: ",score,"/4")
print("Your Score Percentage: ", score/4*100,"%")
```

```python
class Person:
    def __init__(self, name, age):
        self.name=name
        self.age=age
        
```

```python
details=Person("Ade",22)
```

```python
details.name
```

```python
details.age
```

### OBJECT METHODS

Objects can also contain methods and methods in objects are functions that belong to the object.

```python
#Insert a function that prints a greeting and execute it on the details object

class Person:
    def __init__(self, name, age):
        self.name=name
        self.age=age
        
    def greetings(self):
        print("Hello " + self.name)
        
        
```

```python
details=Person("Chris",34)
```

```python
details.greetings()
```

```python

class Person:
    def __init__(inherit, name, age):
        inherit.name=name
        inherit.age=age
        
    def greetings(inherit):
        print("Hello " + inherit.name)
```

```python
details=Person("Ayo",24)
```

```python
details.greetings()
```

```python
#modify object properties

details.name= "Clement"
```

```python
details.greetings()
```

```python
# Delete Object Properties use del keyword

del details.age
```

```python
details.age
```

```python
details.age=34
```

### PYTHON INHERITANCE

Inheritance allows you to define a class that inherits all the methods and properties from another class

- Parent Class : the class being inherited from
- Child Class: the class that inherits from another class

```python
# Create a Parent class

class Person:
    def __init__(self, fname, lname):
        self.fname=fname
        self.lname=lname
        
    def printname(self):
        print(self.fname, self.lname)
```

```python
x=Person("Mathew", "Ani")
x.printname()
```

### create a child class
- send the the parent class as a parameter when creating a child class so it can inherit its properties and methods

```python
class student(Person):
    pass
```

```python
x=student("Jack","Dorsey")
```

```python
x.printname()
```
